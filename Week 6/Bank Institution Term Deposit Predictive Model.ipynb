### EDA to help you understand the data, and get insight about the data generation process 

#### 1. Univariate Analysis on all columns ( You can make use of bar charts, pie charts in understanding the categorical columns. Plots such as histograms can be useful for continuous columns. Be creative and think about better graphs to use to communicate what is in the data)

```python
plt.figure(figsize=(5,5))
sns.countplot(x='y', data=bank_additional_full)
plt.xlabel("Y")
plt.ylabel("Counts")
plt.title("Has the client subscribed to a term deposit? (binary: 'yes','no')")
```

#### This shows that the number of people who have not subscribed to term deposit ismore than the number of people who have.

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age', data=bank_additional_full)
plt.xlabel("Ages")
plt.ylabel("Counts")
plt.title("Counts of Ages in the Dataset")
```

#### This shows that people with age (31) had the most number of counts.

```python
plt.figure(figsize=(5,5))
sns.countplot(x='marital', data=bank_additional_full)
plt.xlabel("Marital Status")
plt.ylabel("Counts")
plt.title("Marital Status Counts")
```

#### This shows that the married people responded than other categories.

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18,6))

bank_additional_full["education"].value_counts().plot(ax=ax[0], kind='bar')
pd.crosstab(bank_additional_full.education, bank_additional_full.y).plot( ax=ax[1],kind='bar')
plt.xlabel("Education Status")
plt.ylabel("Counts")
plt.title("Education Status Counts")
plt.show()
```

### 2. Bivariate Analysis. Carry out some analysis to see how two columns are related. Given that this is a classification problem, cross-tabulation between most categorical columns and the target would be helpful. For numerical columns use boxplots to understand the data separation. 

```python
plt.figure(figsize=(10,5))
plot = sns.boxplot(x = "education", y = "age", hue = "y", data= bank_additional_full)
plt.xticks( rotation=45, horizontalalignment='right' )
plot.set_title('Box plot of Education and Age')
plt.xlabel("Education")
plt.ylabel("Age")

```
